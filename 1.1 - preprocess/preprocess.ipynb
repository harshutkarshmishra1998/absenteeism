7-Extract Weekdays

```python
df['Date'][0].weekday()
# 0 = Monday and 6 = Sunday
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df['Day of the Week'] = df['Date'].apply(date_to_weekday)
```
